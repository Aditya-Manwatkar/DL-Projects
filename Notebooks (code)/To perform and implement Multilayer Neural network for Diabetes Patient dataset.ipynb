###  *To perform and implement Multilayer Neural network for Diabetes Patient dataset*

```python
#importing necessary libraries
from keras.models import Sequential
from keras.layers import Dense 
from sklearn.model_selection import train_test_split
import numpy
import h5py
import pandas as pd
```

```python
#Loading the dataset
data=pd.read_csv("diabetes.csv")
```

```python
data.head(5)
```

### Seggregating dependent and independent variables

```python
X=data.iloc[:,:-1].values
```

```python
Y=data.iloc[:,-1].values
```

```python
X
```

```python
Y
```

### *Splitting dataset into training and validation data*

```python
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=5)
```

### *Build the neural network model*

```python
model=Sequential()
```

```python
model.add(Dense(10,input_dim=8,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(5,activation='relu'))
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=1000,batch_size=10,validation_data=(x_test,y_test))
```

### *Evaluating the models accuracy*

```python
scores=model.evaluate(X,Y)
print("\n%s : %.2f%%" % (model.metrics_names[1],scores[1]*100))
```
